```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
df = pd.read_csv('train.csv')
```

```python
df.columns
```

```python

# Perform exploratory data analysis (EDA)
# Check data types and missing values
print(df.dtypes)
print(df.isnull().sum())

# Check statistical properties of the data
print(df.describe())

# Preprocess the data
# Handle missing values
df.fillna(df.mean(), inplace=True)

# Encode categorical variables
df = pd.get_dummies(df)

# Split the data into training and testing sets
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Build a Ridge regression model
ridge = Ridge()
ridge_params = {'alpha': np.logspace(-4, 0, 100)}
ridge_cv = GridSearchCV(ridge, ridge_params, cv=5)
ridge_cv.fit(X_train, y_train)
y_pred_ridge = ridge_cv.predict(X_test)

# Evaluate the model's performance on the testing set
print("Ridge Regression")
print("Best alpha value:", ridge_cv.best_params_['alpha'])
print("MSE: {:.2f}".format(mean_squared_error(y_test, y_pred_ridge)))
print("R^2 Score: {:.2f}".format(r2_score(y_test, y_pred_ridge)))

# Build a Lasso regression model
lasso = Lasso()
lasso_params = {'alpha': np.logspace(-4, 0, 100)}
lasso_cv = GridSearchCV(lasso, lasso_params, cv=5)
lasso_cv.fit(X_train, y_train)
y_pred_lasso = lasso_cv.predict(X_test)

# Evaluate the model's performance on the testing set
print("\nLasso Regression")
print("Best alpha value:", lasso_cv.best_params_['alpha'])
print("MSE: {:.2f}".format(mean_squared_error(y_test, y_pred_lasso)))
print("R^2 Score: {:.2f}".format(r2_score(y_test, y_pred_lasso)))

# Determine the significant variables and their impact on the target variable
ridge_coefs = pd.DataFrame({'Variable': X.columns, 'Coef': ridge_cv.best_estimator_.coef_})
lasso_coefs = pd.DataFrame({'Variable': X.columns, 'Coef': lasso_cv.best_estimator_.coef_})

print("\nRidge Regression Coefficients:")
print(ridge_coefs.sort_values('Coef', ascending=False))
print("\nLasso Regression Coefficients:")
print(lasso_coefs.sort_values('Coef', ascending=False))
```
